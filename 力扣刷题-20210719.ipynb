```python
# 1838. 最高频元素的频数
# https://leetcode-cn.com/problems/frequency-of-the-most-frequent-element/
from typing import List
def maxFrequency(nums: List[int], k: int) -> int:
#     print(nums,k,end=",")
    nums.sort()
    # 果然是排序加滑动窗口,想到了大概的想法，但是具体实现还是有点问题

    # 想到了滑动窗口，确写不出来？？？太蠢了，
    # 看看别人写的思路
    '''
    刚开始，这题有点难，嗯用例范围10^5，放弃多重循环的暴力吧.
    看到示例3，没毛病不管会不会，至少先排个序吧。
    在瞅示例1吧，能考虑前缀和不，感觉也不对啊。拿[1, 2, 4]来说，1-2是1，2-4是2，那1-4不是还要加一次2。等等，还要加一次？
    这不就是每次右移一位，需要增加(right - left) * (nums[right] - nums[right -1])这么多数字。
    那如果pre_sum不够了，需要收缩左边界呢？只用nums[right] - nums[left]删除掉这个数字，想到这里解题基本就成型了。
    标准的划窗没跑了！我们使用一个滑动窗口模板，默认窗口拉伸，当pre_sum > k时，收缩左边界。直至满足条件，然后每次计算最大窗口距离。
    
    # 滑动窗口模板
    left,right = 0, (0 or 1)
    ret = total = 0
    while right < len(nums):
       更新total值
       while 窗口内数据不满足要求
          1. 更新total值
          2. 收缩左边界
       更新ret最大值
    返回 ret
    
    '''
    
    n = len(nums)
    preSum = 0
    l,r = 0, 1
    ans = 1
    while r < n:
        preSum += (r-l)*(nums[r]-nums[r-1])
        while preSum > k:
            preSum -= nums[r]-nums[l]
            l += 1
        ans = max(ans,r-l+1)
        r += 1
    return ans
    
nums = [1]
k = 5
maxFrequency(nums,k)
```

```python
# 再做一题
# 1647. 字符频次唯一的最小删除次数
def minDeletions(s: str) -> int:
    """
    以后做题都要写思路：
    1、统计所有的频数
    2、对频数再次计算频数
    3、频数进行排序
    得到的字典是这种vv={2: 6, 4: 2, 1: 1, 5: 1}
    代表频数2出现了6次，频数4出现了两次,以此类推
    
    把vv的key进行降序排列
    
    具体的思路是这样
    比如频数4出现2次，频数2出现2次
    
    那么4-2之间有 4-2+1 = 1个空档频数值可以调整，那么我可以把4调整到3，最多调整1个值，还有剩下的我全部记录到下一个频数值也就是2之中。以此类推
    
    
    从出现的频数从高到低排序遍历vv
    i 为当前频数 
    keys[i]-keys[i-1]-1 是相邻的频数之间最多处理多少个重复的频数值
    if keys[i]-keys[i-1] >= vv[keys[i]] -1 意思是相邻的两个频数之间可以处理掉keys[i]的重复数,然后可以计算出需要的代价
    else 把多余的频数值全部调整到keys[i+1] 计算出代价
    
    然后一直遍历就行
    
    最后单独处理一些最后一个值就好了。。
    
    写了好久。。。。。
    
    
    """
    import collections
    myCounter = collections.Counter(s)
    v = list(myCounter.values())
    vv = collections.Counter(v)
    
    keys = list(vv.keys())
    ans = 0
    keys.sort(reverse=True)
#     print(vv)
#     print(keys)
    
    for i in range(0,len(keys)-1):
        frequency = keys[i]  # 当前频数
        if vv[keys[i]] == 1:
            continue
        x = frequency - keys[i+1] -1 # 能解决这么多个频数重复的问题
        
        if x >= vv[keys[i]]-1:
            ans += (vv[keys[i]]-1)*vv[keys[i]]//2
        else:
            t = vv[keys[i]]-1 -x  # 还剩下这么多数需要解决重复
            vv[keys[i+1]] += t
            ans += (1+x)*x//2 + t * (keys[i] - keys[i+1])


    last = keys[-1]
    if vv[last] == 1:
        return ans
    else:
        if vv[last] <= last:
            ans += vv[last]*(vv[last]-1)//2
        else:
            ans += (last-1)*last//2 + (vv[last]-last)*(last)
    print(ans)
        
    pass
s = "aabbccddeeffgttttppppooooo"
minDeletions(s)
```
