```python
# LCP 07. 传递信息
# https://leetcode-cn.com/problems/chuan-di-xin-xi/

# 采用BFS
from typing import List
def numWays(n: int, relation: List[List[int]], k: int) -> int:
    # 经过k轮传到n-1
    
    # 天天BFS，我真的菜啊，代码写的极差,真的dfs
    k1 = []
    ans = []
    
    def dfs(relations,k):
        if k == 0:
            for r in relations:
                if r[1] == n-1:
                    ans.append(1)
            return 
#         print(f"relations = {relations}")
        for r in relations:
            x =[]
            for rr in relation:
                print(r,rr)
                if r[1] == rr[0]:
                    x.append(rr)
#             print(f"x= {x}")
            dfs(x,k-1)
        
    x = []
    for rr in relation:
        if rr[0]== 0:
            x.append(rr)
    dfs(x,k-1)
    print(sum(ans))
    pass
n = 5
relation = [[0,2],[2,1],[3,4],[2,3],[1,4],[2,0],[0,4]]
k = 3
numWays(n,relation,k)
```
