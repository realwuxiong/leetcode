```python
# 494. 目标和
# https://leetcode-cn.com/problems/target-sum/


# 用回溯法试试
from typing import List
def findTargetSumWays(nums: List[int], target: int) -> int:
    stack = []
    ans = []
    def dfs(i):
        if i > len(nums):
            if sum(stack) == target:
                ans.append(1)
            return 
#         print(i)
        for j in range(i,len(nums)):
            stack.append(nums[j]*1)
            dfs(j+1)
            stack.pop()
            stack.append(nums[j]*-1)
            dfs(j+1)
            stack.pop()
    dfs(0)
    print(sum(ans))
    pass
nums = [1,1,1,1,1]
target = 3
findTargetSumWays(nums,target)
```
