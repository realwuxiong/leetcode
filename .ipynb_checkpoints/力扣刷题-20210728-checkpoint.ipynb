```python
# 863. 二叉树中所有距离为 K 的结点
# https://leetcode-cn.com/problems/all-nodes-distance-k-in-binary-tree/

# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None


class Solution:
    def distanceK(self, root: TreeNode, target: TreeNode, k: int) -> List[int]:
        # 这个题目有点意思，哈希表，反过来找父节点遍历。用frm判断是否遍历过，
        # 标准答案写的确实漂亮啊
        # 用dfs
        
        # 我明白了，需要用hash表记录父节点
        myHash = {}
        ans = []
        myHash[root.val] = None
        def findParents(root):
            if root == None:
                return 
            if  root.left != None:   
                myHash[root.left.val] = root
                findParents(root.left)
            if  root.right != None:
                myHash[root.right.val] = root
                findParents(root.right)
        def findSons(node,frm,t,k):
            if not node:
                return 
            if t == k:
                ans.append(node.val)
                return 

            # 搜索左子树
            if node.left != frm:
                findSons(node.left,node,t+1,k)
            # 搜索右子树
            if node.right != frm:
                findSons(node.right,node,t+1,k)
            # 搜索父节点
            if myHash[node.val] != frm:
                findSons(myHash[node.val],node,t+1,k)

        findParents(root)
        # print(myHash)
        findSons(target,None,0,k)
        return ans

            
```
