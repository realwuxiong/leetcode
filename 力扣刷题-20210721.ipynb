```python
# 31. 下一个排列
# https://leetcode-cn.com/problems/next-permutation/
# 再来一个题
from typing import List
def nextPermutation(nums: List[int]) -> None:
#     print(nums)
    # 两遍扫描，看答案，然后自己尝试写一下这个代码
    n = len(nums)
    i = n-2
    while i > 0:
        if nums[i] < nums[i+1]:
            break
        i -= 1
    
    # 然后找j
    j = n-1
    while j > i:
        if nums[j] > nums[i]:
            nums[i],nums[j] = nums[j],nums[i]
            break
        j -= 1
    # i后面的全部翻转
    low = 0 if i == 0 and j == 0 else i+1 
    high = n-1
    while low < high:
#         print(low,high)
        nums[low],nums[high] = nums[high],nums[low]
        low += 1
        high -= 1
#     print(nums)
    return nums
    
    pass
nums = [1,2,3]
nextPermutation(nums)
```
